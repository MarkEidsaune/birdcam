```python
import json
import os 
import requests
import time

subscription_key = '9149d9456c8246e88ddc9de23559d151'
endpoint = 'https://api.bing.microsoft.com/v7.0/images/search'
q = 'Florida Scrub-Jay'
count = 150
offset = 0
total_imgs = 0
est_matches = 1000
max_imgs = 10000
content_urls = []

while (total_imgs <= est_matches) or (total_imgs <= max_imgs):
    print(f'offset: {offset}, count: {count}, matches: {est_matches}')
    headers = { 'Ocp-Apim-Subscription-Key': subscription_key }
    params = {
            'q': q,
            'count': count,
            'offset': offset
        }
    try:
        r = requests.get(endpoint, headers=headers, params=params)
        r.raise_for_status()
    except Exception as ex:
        raise ex
        
    result = r.json()
    
    for img in result['value']:
        content_urls.append(img['contentUrl'])
    
    matches = result['totalEstimatedMatches']
    offset = result['nextOffset']
    total_imgs += len(content_urls)
    
    print(f'\tEstimated matches: {matches}')
    print(f'\tTotal image urls: {total_imgs}')
    
    time.sleep(1)
```

```python
output_dir = '/media/nvme2/birds/raw/'

for i, url in enumerate(content_urls[:50]):
    q_formatted = q.replace(" ", "_").lower()
    fname = f'{q_formatted}_{i}.jpg'
    print(f'Saving {fname}')
    path = os.path.join(output_dir, fname)
    try:
        with open(path, "wb") as f:
            image_data = requests.get(url)
            f.write(image_data.content)
    except OSError as e:
        raise e
```
